```python
import requests
import json

# AWS Pricing API (Filtered Query)
def get_filtered_aws_pricing():
    url = "https://api.pricing.us-east-1.amazonaws.com"
    headers = {
        'Content-Type': 'application/json'
    }
    
    # Define filters for EC2 instance type and region
    params = {
        "serviceCode": "AmazonEC2",
        "filters": [
            {
                "field": "instanceType",
                "value": "t3.micro",  # Specify instance type
                "type": "TERM_MATCH"
            },
            {
                "field": "location",
                "value": "US East (N. Virginia)",  # Specify AWS region
                "type": "TERM_MATCH"
            }
        ],
        "maxResults": 10  # Limit results to avoid large payloads
    }
    
    # Make a POST request to the AWS pricing API
    response = requests.post(f"{url}/v1.0/aws/AmazonEC2/current/priceList", json=params, headers=headers)
    
    if response.status_code == 200:
        data = response.json()
        return data
    else:
        print(f"Failed to retrieve AWS pricing data: {response.status_code}")
        return {}

# Google Cloud Pricing (Smaller JSON Dataset)
def get_gcp_pricing():
    url = "https://cloudpricingcalculator.appspot.com/static/data/pricelist.json"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        gcp_pricing = []
        for sku, pricing_info in data.items():
            if isinstance(pricing_info, dict) and "description" in pricing_info:
                if "Compute" in pricing_info.get("category", ""):
                    gcp_pricing.append({
                        "Description": pricing_info.get("description", "N/A"),
                        "Price": pricing_info.get("pricing", {})
                    })
        return gcp_pricing
    else:
        print(f"Failed to retrieve GCP pricing data: {response.status_code}")
        return []

# Azure Pricing API (No API Key Required)
def get_azure_pricing():
    url = "https://prices.azure.com/api/retail/prices"
    params = {
        "$filter": "serviceName eq 'Virtual Machines'"
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        azure_pricing = []
        for item in data.get("Items", []):
            azure_pricing.append({
                "Instance Type": item.get("armSkuName", "N/A"),
                "Location": item.get("armRegionName", "N/A"),
                "Price": item.get("retailPrice", "N/A"),
                "OS": item.get("productName", "N/A")
            })
        return azure_pricing
    else:
        print(f"Failed to retrieve Azure pricing data: {response.status_code}")
        return []

# Main function to fetch pricing data from AWS, GCP, and Azure
def get_cloud_pricing():
    print("Fetching filtered AWS Pricing...")
    aws_pricing = get_filtered_aws_pricing()
    
    print("Fetching Google Cloud Pricing...")
    gcp_pricing = get_gcp_pricing()
    
    print("Fetching Azure Pricing...")
    azure_pricing = get_azure_pricing()

    return {
        "AWS": aws_pricing,
        "Google Cloud": gcp_pricing,
        "Azure": azure_pricing
    }

# Save the results to JSON files
if __name__ == "__main__":
    cloud_pricing = get_cloud_pricing()
    
    # Save AWS pricing data
    with open('filtered_aws_pricing.json', 'w') as f:
        json.dump(cloud_pricing['AWS'], f, indent=4)

    # Save GCP pricing data
    with open('gcp_pricing.json', 'w') as f:
        json.dump(cloud_pricing['Google Cloud'], f, indent=4)

    # Save Azure pricing data
    with open('azure_pricing.json', 'w') as f:
        json.dump(cloud_pricing['Azure'], f, indent=4)

    print("Cloud pricing data saved to JSON files.")
```
