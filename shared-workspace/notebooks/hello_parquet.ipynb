```python
import os
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

# Step 1: Create a simple Pandas DataFrame
data = {
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35]
}
df = pd.DataFrame(data)

# Step 2: Convert the DataFrame to an Apache Arrow Table
table = pa.Table.from_pandas(df)

# Step 3: Define the custom path where you want to save the Parquet file
directory = "/home/jovyan/parquet"
filename = "hello_world.parquet"
file_path = os.path.join(directory, filename)

# Ensure the directory exists
os.makedirs(directory, exist_ok=True)

# Step 4: Write the Arrow Table to a Parquet file at the specified path
pq.write_table(table, file_path)

print(f"Parquet file created: {file_path}")
```
