```c++
#include <iostream>
#include <string>

// Print a welcome message and flush the output
std::cout << "Hello, World! Welcome to C++ in Jupyter Notebook." << std::endl;

// Define and print a variable
int number = 5;
std::cout << "The value of the variable 'number' is: " << number << std::endl;

// Simulate user input (since std::cin may not work as expected in notebooks)
std::string name = "JupyterUser";
std::cout << "Hello, " << name << "!" << std::endl;

// Simple loop example with output
std::cout << "Counting from 1 to 5:" << std::endl;
for (int i = 1; i <= 5; i++) {
    std::cout << i << std::endl;
}

// Flush output to ensure everything is displayed
std::cout << std::flush;
```
